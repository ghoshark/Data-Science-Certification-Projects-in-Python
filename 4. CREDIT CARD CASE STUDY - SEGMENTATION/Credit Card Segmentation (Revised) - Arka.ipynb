### K-means Clustering

```python
from sklearn.cluster import KMeans
```

```python
distortions=[]
K=range(1,10)
for k in K:
    kmean_model=KMeans(n_clusters=k)
    kmean_model.fit(df2)
    distortions.append(kmean_model.inertia_)
```

```python
distortions
```

```python
plt.plot(K,distortions)
```

### Silhouette Coefficient

```python
# calculate SC for K=3 through K=12
k_range = range(3, 12)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(df2)
    scores.append(silhouette_score(df2, km.labels_))
```

```python
scores
```

```python
# plot the results
plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

#### As per Silhouette Analysis, the optimal number of clusters can be 4,5 or 6

### Elbow Analysis

```python
cluster_range = range( 1, 20 )
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit( df2 )
    cluster_errors.append( clusters.inertia_ )
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )

clusters_df[0:10]
```

```python
# allow plots to appear in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

### Note:

- The elbow diagram shows that the gain in explained variance reduces significantly between 3 and 5. So, optimal number of clusters could either 4 or 5.

- The actual number of clusters chosen can be finally based on business context and convenience of dealing with number of segments or clusters.
